## 1. Lists

A **list** is an ordered, changeable collection of items. Lists are written with square brackets `[ ]`, and items are separated by commas.

Lists are useful whenever you need to keep track of a sequence of things — for example, the scores of every student in a class, or the words in a sentence.


````python
# Creating a list
fruits = ["apple", "banana", "cherry"]
print(fruits)
print(type(fruits))
````

### Accessing items

Items in a list are accessed by their **index**, starting at 0.


````python
print(fruits[0])   # first item
print(fruits[-1])  # last item
print(fruits[1:])  # everything from index 1 onward (a "slice")
````

### Modifying a list

Lists are *mutable*, meaning you can change them after creating them.


````python
fruits.append("date")       # add an item to the end
fruits[0] = "avocado"        # replace an item
fruits.remove("cherry")      # remove a specific item
print(fruits)
print("Number of fruits:", len(fruits))
````

### Looping through a list

Very often you will want to do something with every item in a list.


````python
for fruit in fruits:
    print(f"I like {fruit}")
````

## 2. Tuples

A **tuple** looks like a list, but it is written with parentheses `( )` and, once created, **cannot be changed** (it is *immutable*). Tuples are great for grouping related values that should not change, like coordinates or a (year, month, day) date.


````python
point = (3, 4)
print(point)
print("x =", point[0], " y =", point[1])

# This would raise an error if uncommented, because tuples are immutable:
# point[0] = 10
````

Tuples are commonly used to return more than one value from a function.


````python
def min_max(numbers):
    return (min(numbers), max(numbers))

result = min_max([4, 9, 1, 7])
print(result)

low, high = result   # "unpacking" a tuple into two variables
print("low:", low, " high:", high)
````

## 3. Dictionaries

A **dictionary** stores data as `key: value` pairs, written with curly braces `{ }`. Instead of looking items up by position (like a list), you look them up by a meaningful key.

Dictionaries are perfect for representing a single "record," such as one student's information.


````python
student = {
    "name": "Wesley",
    "major": "Computer Science",
    "year": 1
}

print(student)
print(student["name"])
````

### Adding, updating, and removing keys


````python
student["gpa"] = 3.8          # add a new key
student["year"] = 2            # update an existing key
del student["major"]           # remove a key
print(student)
````

### Looping through a dictionary


````python
for key, value in student.items():
    print(f"{key}: {value}")
````

## 4. Sets

A **set** is an unordered collection of **unique** items, written with curly braces `{ }` (but without `key: value` pairs like a dictionary). Sets automatically remove duplicates and are great for membership checks and comparing groups of items.


````python
a = {1, 2, 3, 3, 2}
print(a)   # duplicates are removed automatically

b = {3, 4, 5}
print("Union:", a | b)          # items in either set
print("Intersection:", a & b)   # items in both sets
print("Difference:", a - b)     # items in a but not b
print("Is 2 in a?", 2 in a)
````

## 5. Working with Files

So far every piece of data has lived only in memory and disappeared once the notebook stopped running. **Files** let you save data permanently and load it back later.

The standard way to work with a file in Python is the `with open(...) as f:` pattern. It automatically closes the file for you when you're done, even if an error occurs.


````python
# Writing to a file
lines_to_write = ["apple\n", "banana\n", "cherry\n"]

with open("fruits.txt", "w") as f:
    f.writelines(lines_to_write)

print("File written!")
````

### Reading a file

There are a few common ways to read a file:


````python
# Read the whole file as one string
with open("fruits.txt", "r") as f:
    content = f.read()
print(repr(content))
````

````python
# Read the file line by line into a list
with open("fruits.txt", "r") as f:
    lines = f.readlines()
print(lines)

# Often you'll want to strip the newline characters
clean_lines = [line.strip() for line in lines]
print(clean_lines)
````

````python
# Loop directly over a file, one line at a time (memory-efficient for big files)
with open("fruits.txt", "r") as f:
    for line in f:
        print("Fruit:", line.strip())
````

### Appending to a file

Opening a file with mode `"a"` (append) adds new content to the end instead of overwriting it.


````python
with open("fruits.txt", "a") as f:
    f.write("date\n")

with open("fruits.txt", "r") as f:
    print(f.read())
````

## 6. Data Processing: Putting It All Together

Real programs usually combine everything you've learned so far: read data from a file into a data structure, process it, and maybe write the results back out.

Let's work through a small example: a CSV-style file of student grades.


````python
# First, let's create a sample data file to work with
csv_data = """name,score
Ana,88
Bruno,72
Carla,95
Diego,64
Elena,79
"""

with open("grades.csv", "w") as f:
    f.write(csv_data)

print("grades.csv created!")
````

### Parsing the file manually

We can split each line on the comma to turn the text into a list of dictionaries.


````python
students = []

with open("grades.csv", "r") as f:
    lines = f.readlines()

header = lines[0].strip().split(",")   # ['name', 'score']

for line in lines[1:]:
    name, score = line.strip().split(",")
    students.append({"name": name, "score": int(score)})

print(students)
````

### Processing the data

Now that the data lives in a list of dictionaries, we can use everything from Sections 1–4 to analyze it.


````python
scores = [s["score"] for s in students]

print("Average score:", sum(scores) / len(scores))
print("Highest score:", max(scores))
print("Lowest score:", min(scores))

passing = [s["name"] for s in students if s["score"] >= 70]
print("Students who passed (>= 70):", passing)
````

### Using the built-in `csv` module

Python's standard library also has a `csv` module that handles parsing details (like commas inside quoted text) for you. It's the recommended way to work with real CSV files.


````python
import csv

with open("grades.csv", "r") as f:
    reader = csv.DictReader(f)
    students_csv = [row for row in reader]

print(students_csv)
# Note: csv.DictReader reads every value as a string, so scores need to be converted
for s in students_csv:
    s["score"] = int(s["score"])
print(students_csv)
````

### Writing processed results back to a file

Finally, let's write a summary report out to a new file.


````python
with open("summary.txt", "w") as f:
    f.write(f"Average score: {sum(scores) / len(scores):.2f}\n")
    f.write(f"Highest score: {max(scores)}\n")
    f.write(f"Lowest score: {min(scores)}\n")
    f.write("Passing students:\n")
    for name in passing:
        f.write(f"  - {name}\n")

with open("summary.txt", "r") as f:
    print(f.read())
````

## Summary

| Structure / Tool | Ordered? | Mutable? | Duplicates? | Typical use |
|---|---|---|---|---|
| `list` | Yes | Yes | Yes | A sequence of items |
| `tuple` | Yes | No | Yes | A fixed group of related values |
| `dict` | Yes (insertion order) | Yes | Keys unique | A record of key/value pairs |
| `set` | No | Yes | No | Unique items, membership tests |
| Files (`open`) | — | — | — | Persisting and loading data |
